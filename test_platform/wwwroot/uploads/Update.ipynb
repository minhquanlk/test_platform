```python
import pandas as pd
import numpy as np
import json
file = open('quizapp-d066c-default-rtdb-export (24).json', encoding="utf8")

data = json.load(file)
list_hist = list(data['History'].keys())
list_log = list(data['LoginLog'].keys())
list_start = list(data['Start'].keys())
list_candidates_hist = [s for s in list_hist if s.startswith('-')]
list_candidates_log = [s for s in list_log if s.startswith('-')]
list_candidates_start = [s for s in list_start if s.startswith('-')]
filtered_dict_list_log = {key: value for key, value in data['Candidates'].items() if key in list_candidates_log}
filtered_dict_list_start = {key: value for key, value in data['Candidates'].items() if key in list_candidates_start}
filtered_dict_list_not_start = {key: value for key, value in data['Candidates'].items() if key not in list_candidates_start}
filtered_dict_list_start_time = {key: value for key, value in data['Start'].items() if key in list_candidates_start}
filtered_dict_list_hist = {key: value for key, value in data['Candidates'].items() if key in list_candidates_hist}
filtered_dict_data = {key: value for key, value in data['Candidates'].items()}
df_log = pd.DataFrame.from_dict(filtered_dict_list_log, orient='index')
df_start = pd.DataFrame.from_dict(filtered_dict_list_start, orient='index')
df_not_start = pd.DataFrame.from_dict(filtered_dict_list_not_start, orient='index')
df_time = pd.DataFrame.from_dict(filtered_dict_list_start_time, orient='index')
df_hist = pd.DataFrame.from_dict(filtered_dict_list_hist, orient='index')
df_data = pd.DataFrame.from_dict(filtered_dict_data, orient='index')

```

```python
with pd.ExcelWriter('update16h32_28112023.xlsx') as writer:
    df_log.to_excel(writer, sheet_name='Login', index=True)
    df_start.to_excel(writer, sheet_name='Start', index=True)
    df_not_start.to_excel(writer, sheet_name='Not Start', index=True)
    df_hist.to_excel(writer, sheet_name='Submit', index=True)
    df_time.to_excel(writer, sheet_name='Time', index=True)
    df_data.to_excel(writer, sheet_name='Data', index=True)
```

```python
list_correct_ans = {key: "a" for key in data['TestQuestion'].keys()}
for item in list_correct_ans.items():
    list_correct_ans[item[0]] = data['TestQuestion'][item[0]]['CorrectAnswer']['ans1']
    
```

```python
df_hist['Điểm Numberical'] = 0
df_hist['Điểm Logical'] = 0
df_hist['Điểm Problem Solving'] = 0
df_hist['Điểm Critical Thinking'] = 0
df_hist['Tổng điểm'] = 0
df_hist
```

```python
items_list = []
list_candidate_submit = {key: "a" for key in data['History'].keys()}
for item in list_candidate_submit.items():
    score_numberical = 0
    score_logical = 0
    score_problem = 0
    score_critical = 0
    list_candidate_submit[item[0]] = data['History'][item[0]]['UserChoices']
    items_list = list(list_candidate_submit[item[0]].items())

    for i in range(0, 15):
        if (list_correct_ans[items_list[i][0]] == items_list[i][1]['ans']):
            score_numberical+=1
            
    for i in range(15, 30):
        if (list_correct_ans[items_list[i][0]] == items_list[i][1]['ans']):
            score_logical+=1
    
    for i in range(30, 34):
        if (list_correct_ans[items_list[i][0]] == items_list[i][1]['ans']):
            score_problem+=1
    
    for i in range(34, 40):
        if (list_correct_ans[items_list[i][0]] == items_list[i][1]['ans']):
            score_critical+=1
    df_hist.at[item[0], 'Điểm Numberical'] = score_numberical
    df_hist.at[item[0], 'Điểm Logical'] = score_logical
    df_hist.at[item[0], 'Điểm Problem Solving'] = score_problem
    df_hist.at[item[0], 'Điểm Critical Thinking'] = score_critical
    df_hist.at[item[0], 'Tổng điểm'] = score_numberical + score_logical + score_problem + score_critical
# list_candidate_submit['-NgshAu_dnK4hekiZ0kY']
```

```python
with pd.ExcelWriter('update18h15_28112023_result.xlsx') as writer:
    df_log.to_excel(writer, sheet_name='Login', index=True)
    df_start.to_excel(writer, sheet_name='Start', index=True)
    df_not_start.to_excel(writer, sheet_name='Not Start', index=True)
    df_hist.to_excel(writer, sheet_name='Submit', index=True)
    df_time.to_excel(writer, sheet_name='Time', index=True)
    df_data.to_excel(writer, sheet_name='Data', index=True)
```
